#### Problem Statement:
The excel file contains images of two feet or rather high values in a pattern that represent a foot. What you need to do is extract the positions and the values of the edges of the feet. You can do this by assuming both the feet as one image and extract the left and right edges of the entire image or you can isolate each foot and extract their edges individually.

#### Importing important libraries

```python
import openpyxl 
import pandas as pd 
```

#### Operations 

```python
# Give the location of the file 
path = r'C:\Users\YASHA\Desktop\Assignment01_1\Assignment01\img.xlsx'
  
# workbook object is created 
wb_obj = openpyxl.load_workbook(path) 

# worksheet object is created
sheet_obj = wb_obj.active 
```

#### For Left feet

```python
# lists for appending Final values
rows = []
columns = []
position = []

# looping for structure
for j in range(6, 29): # j represents rows
    for i in range(3,20): # i represents columns
        cell_obj = sheet_obj.cell(row = j,column = i)
        cell_obj_lft = sheet_obj.cell(row = j,column = i - 1)
        cell_obj_rt = sheet_obj.cell(row = j,column = i + 1)
        if ((cell_obj.value is not None) & (cell_obj_lft.value is None)) or ((cell_obj.value is not None) & (cell_obj_rt.value is None)):
            rows.append(cell_obj.row)
            columns.append(cell_obj.column)
            position.append(cell_obj.value)
                        
# initialize list of lists 
data = [rows, columns, position] 
  
# Create the pandas DataFrame 
Left_feet = pd.DataFrame(data, index  =['lft_rows', 'lft_columns','lft_position']) 
  
# print dataframe. 
Left_feet.T
```

#### For right feet

```python
# lists for appending Final values
rows = []
columns = []
position = []

# looping for structure
for j in range(6, 29): # j represents rows
    for i in range(21,35): # i represents columns
        cell_obj = sheet_obj.cell(row = j,column = i)
        cell_obj_lft = sheet_obj.cell(row = j,column = i - 1)
        cell_obj_rt = sheet_obj.cell(row = j,column = i + 1)
        if ((cell_obj.value is not None) & (cell_obj_lft.value is None)) or ((cell_obj.value is not None) & (cell_obj_rt.value is None)):
            rows.append(cell_obj.row)
            columns.append(cell_obj.column)
            position.append(cell_obj.value)
                        
# initialize list of lists 
data = [rows, columns, position] 
  
# Create the pandas DataFrame 
Right_feet = pd.DataFrame(data, index  =['rt_rows', 'rt_columns','rt_position']) 
  
# print dataframe. 
Right_feet.T
```
